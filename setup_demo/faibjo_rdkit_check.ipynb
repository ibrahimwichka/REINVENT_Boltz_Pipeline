```python
"""Practice notebook for the chemical analysis setup tutorial.
Computes Murcko scaffolds and QED for a few test molecules."""

from rdkit import Chem
from rdkit.Chem import QED
from rdkit.Chem.Scaffolds import MurckoScaffold

ACCESS_CODE = "faibjo"  # replace with your own 3+3 code

test_smiles = [
    "CCO",
    "c1ccccc1",
    "CC(=O)Oc1ccccc1C(=O)O",
    "CN1C=NC2=C1C(=O)N(C(=O)N2C)C",
]
lines = [f"# Chemical analysis environment check -- {ACCESS_CODE}", ""]
for smi in test_smiles:
    mol = Chem.MolFromSmiles(smi)
    scaffold = MurckoScaffold.MurckoScaffoldSmiles(mol=mol)
    qed_score = QED.qed(mol)
    lines.append(f"{smi:<30} scaffold={scaffold:<20} QED={qed_score:.3f}")

output_path = f"{ACCESS_CODE}_rdkit_results.txt"
with open(output_path, "w") as f:
    f.write("\n".join(lines) + "\n")


print(f"Wrote results to {output_path}")
```
